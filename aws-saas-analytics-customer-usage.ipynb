**DAILY ACTIVE CUSTOMERS (DAC)**

```python
daily_active_customers = fact_usage.groupby('Event_Date')['Customer_ID'].nunique().reset_index(name = 'Daily_Active_Customers')
```

```python
daily_active_customers
```

```python
monthly_active_customers = fact_usage.groupby('Event_Month')['Customer_ID'].nunique().reset_index(name = 'Monthly_Active_Customers')
```

```python
monthly_active_customers
```

**Active Days per Customer**

```python
active_days_per_customer = (
    fact_usage.groupby("Customer_ID")["Event_Date"]
    .nunique()
    .reset_index(name="Active_Days")
)
active_days_per_customer
```

**Usage Frequency per Customer**

```python
event_count_per_customer = (
    fact_usage.groupby("Customer_ID")
    .size()
    .reset_index(name="Total_Events")
)
event_count_per_customer
```

**Event Type Distribution**

```python
event_type_dist = (
    fact_usage.groupby(["Customer_ID", "Event_Type"])
    .size()
    .reset_index(name="Event_Count")
)
event_type_dist
```

**Product Usage**

```python
product_usage = (
    fact_usage.groupby(["Customer_ID", "Product_ID"])
    .size()
    .reset_index(name="Product_Event_Count")
)
product_usage
```

```python
fact_usage["Event_Timestamp"].max().normalize()
```

**Recency (Days Since Last Event)**

```python
max_date = fact_usage["Event_Timestamp"].max()

recency = (
    fact_usage.groupby("Customer_ID")["Event_Timestamp"]
    .max()
    .reset_index()
)

recency["Recency_Days"] = (max_date - recency["Event_Timestamp"]).dt.days

recency
```

**Rolling 7-day / 30-day Frequency**

```python
daily_events = (
    fact_usage
    .groupby(["Customer_ID", "Event_Date"])
    .size()
    .reset_index(name="Events")
)

daily_events["Event_Date"] = pd.to_datetime(daily_events["Event_Date"])
daily_events
```

**For 7 Days**

```python
daily_events = (
    daily_events
    .set_index("Event_Date")
    .groupby("Customer_ID")["Events"]
    .rolling("7D")
    .sum()
    .reset_index()
    .rename(columns={"Events": "Events_Last_7_Days"})
)

daily_events
```

**For 30 Days:**

```python
daily_events_30 = (
    daily_events
    .set_index("Event_Date")
    .groupby("Customer_ID")["Events_Last_7_Days"]
    .rolling("30D")
    .sum()
    .reset_index()
    .rename(columns={"Events_Last_7_Days": "Events_Last_30_Days"})
)

daily_events_30
```

**Customer-Level Health Score**

```python
feature_variety = (
    fact_usage.groupby("Customer_ID")["Event_Type"]
    .nunique()
    .reset_index(name="Event_Type_Variety")
)

feature_variety
```

```python
health = (
    recency[["Customer_ID", "Recency_Days"]]
    .merge(feature_variety, on="Customer_ID", how="left")
    .merge(event_count_per_customer, on="Customer_ID", how="left")
)
```

```python
health["Health_Score"] = (
    (health["Event_Type_Variety"] * 2) + 
    (health["Total_Events"] * 0.1) -
    (health["Recency_Days"] * 0.5)
)

health
```
